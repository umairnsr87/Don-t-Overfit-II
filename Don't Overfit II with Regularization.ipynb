```python
df_train=pd.read_csv("/kaggle/input/dont-overfit-ii/train.csv")

df_test=pd.read_csv("/kaggle/input/dont-overfit-ii/test.csv")
```

```python
#lets drop the index column since it wont be contributing in any sense
df_train.drop(["id"],axis=1,inplace=True)
df_test.drop(["id"],axis=1,inplace=True)
```

```python
#we have saperated target data and the rest of the data
y_train=df_train["target"]
del df_train["target"]

```

```python
from sklearn.preprocessing import MinMaxScaler,RobustScaler,StandardScaler
```

```python
ss=StandardScaler()
rs=RobustScaler()
ms=MinMaxScaler()
```

```python
df_train=ss.fit_transform(df_train)
df_test=ss.fit_transform(df_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
clf.fit(df_train,y_train)
```

```python
clf.predict_proba()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param={
    "penalty":['l1','l2'],
    "solver" : ['liblinear'],
    'class_weight':['balanced', None],
    "max_iter":[100,500,1000,2500,5000,10000]
}
```

```python
Grid=GridSearchCV(param_grid=param,estimator=clf,n_jobs=-1,cv=5,verbose=2)
```

```python
Grid.fit(df_train,y_train)
```

```python
Grid.best_params_,Grid.best_estimator_,Grid.best_score_
```

```python
clf=LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                    intercept_scaling=1, l1_ratio=None, max_iter=100,
                    multi_class='warn', n_jobs=None, penalty='l1',
                    random_state=None, solver='liblinear', tol=0.0001, verbose=0,
                    warm_start=False)
clf.fit(df_train,y_train)
```

```python
predicted=clf.predict(df_test)
```
